> # **Heart Disease Dataset**
>**Features / Target**
>1. **age** – Patient’s age (in years)
>2. **sex** – Patient’s gender
>
>       * 0 = Female
>       * 1 = Male
>
>3. **cp** – Chest pain type
>
>       * 0 = Typical angina
>       * 1 = Atypical angina
>       * 2 = Non-anginal pain
>       * 3 = Asymptomatic
>
>4. **trestbps** – Resting blood pressure (mm Hg)
>5. **chol** – Serum cholesterol (mg/dl)
>6. **fbs** – Fasting blood sugar
>
>       * 0 = ≤ 120 mg/dl
>       * 1 = > 120 mg/dl
>
>7. **restecg** – Resting electrocardiographic results
>
>       * 0 = Normal
>       * 1 = ST-T wave abnormality
>       * 2 = Left ventricular hypertrophy (by Estes’ criteria)
>
>8. **thalach** – Maximum heart rate achieved
>9. **exang** – Exercise-induced angina
>       * 0 = No
>       * 1 = Yes
>
>10. **oldpeak** – ST depression induced by exercise (relative to rest)
>11. **slope** – Slope of the peak exercise ST segment
>
>       * 0 = Upsloping
>       * 1 = Flat
>       * 2 = Downsloping
>
>12. **ca** – Number of major vessels (0–3) colored by fluoroscopy
>13. **thal** – Thalassemia status
>
>       * 3 = Normal
>       * 1 = Fixed defect
>       * 2 = Reversible defect
>
>14. **target** – Presence of heart disease
>
>       * 0 = No heart disease
>       * 1 = Heart disease present
>
>**Objectives**
>1. Age & Heart Disease Risk - Identify how heart disease prevalence changes with age.
>
>2. Gender Differences - Compare male vs. female heart disease rates.
>
>3. Cholesterol Levels (chol) - Understand the relationship between serum cholesterol and heart disease.
>
>4. Blood Pressure (trestbps) - Investigate whether high resting blood pressure correlates with disease presence.
>
>5. Maximum Heart Rate (thalach) - See if people with lower maximum heart rates during exercise are more likely to have heart disease.
>

## Data Visualization, Storytelling & Experimenting with charts

Heart Disease Data count

```python
counts = df["target"].value_counts()
labels = ["Heart Disease", "No Heart Disease"]
colors = ['#5E4B56', '#AFD2E9']
plt.figure(figsize=(5, 5))
plt.pie(counts,autopct='%1.1f%%', colors=colors, wedgeprops={'edgecolor' : 'black'})
plt.title("Heart Disease Distribution", fontsize=13)
plt.legend(labels)
plt.show()
```

Correlation Heatmap between the points of study ['age', 'sex', 'trestbps', 'chol', 'target']

```python
plt.figure(figsize=(8, 5))
corr = df[['age', 'sex', 'trestbps', 'chol', 'target']].corr()
sns.heatmap(corr, annot=corr, cmap="PuBu")
plt.title("Correlation Map")
plt.show()
```

Visualize the Relationship between Age & Heart-disease

```python
colors = ['#AFD2E9', '#5E4B56']
plt.figure(figsize=(8, 5))
sns.histplot(df, x="age", hue="target", bins= 30, kde=True, multiple='dodge', palette=colors)
plt.title("Age - Heart Disease Count")
plt.xlabel("Age", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

Visualize the relationship between Gender & Heart-Disease

```python
plt.figure(figsize=(5, 5))
sns.histplot(df, x="sex", hue="target", multiple='dodge', discrete= True, palette=colors)
plt.xticks([0, 1], labels=['Female', 'Male'])
plt.title("Heart Disease Count by Gender", fontsize=12)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Count", fontsize=12)
sns.despine()
plt.show()
```

Discover the relationship between Cholestrol & Heart Disease

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(df, x="chol", hue="target", fill=True, palette=colors)
plt.title("Cholesterol Level by Heart Disease Status", fontsize= 13)
plt.xlabel("Cholesterol", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

Relationship between Blood Pressure & Heart Disease

```python
plt.figure(figsize=(8, 5))
sns.violinplot(hue="target", y="trestbps", data=df, palette=colors)
plt.title("Heart Diseases by Blood Pressure")
plt.xlabel("Heart Disease")
plt.ylabel("Resting Blood Pressure")
plt.show()
```

Maximum Heart Rate x Heart Disease

```python
plt.figure(figsize=(8, 5))
sns.histplot(df, x="trestbps", hue="target", multiple='dodge', kde=True, palette=colors)
plt.title("Maximum Heart Rate by Heat Disease Status", fontsize= 13)
plt.xlabel("Max-Heart Rate", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

# **Final Insights**

**1. Cholesterol Level by Heart Disease Status (KDE Plot)**
   * **Observation:** Both patients with and without heart disease tend to have cholesterol levels clustering around 200–250 mg/dL.

   * **Notable Detail:** There is substantial overlap between the two groups, meaning cholesterol alone **may not perfectly separate heart disease vs. no heart disease cases**.

   * **Medical Insight:** While elevated cholesterol is a known risk factor, this plot suggests it should be considered alongside other variables (like blood pressure, age, or max heart rate) for better predictive power

**2. Heart Diseases by Blood Pressure (Violin Plot)**
 * **Observation:** Both groups have a similar median resting blood pressure (~130 mmHg), but the distribution is slightly broader in those without heart disease.

 * **Notable Detail:** Both distributions include high blood pressure cases (>160 mmHg), **which is a known cardiovascular risk factor**.

 * **Medical Insight:** Hypertension is prevalent in both groups, but the lack of a large shift in median values suggests that **blood pressure alone may not clearly distinguish between the groups** — though it remains clinically relevant when combined with other indicators.

**3. Maximum Heart Rate by Heart Disease Status (Histogram)**
 * **Observation:** Patients with **heart disease** tend to **have higher maximum heart rates**, with a concentration between 120–140 bpm, whereas those without heart disease are more evenly spread.

 * **Notable Detail:** The higher counts of heart disease cases in the 120–140 bpm range could indicate that these patients reached elevated heart rates under stress (possibly from stress tests).

 * **Medical Insight:** Maximum heart rate response could be a **stronger differentiator between heart disease and non-heart disease patients** than cholesterol or blood pressure alone.
